```python
# Допоміжне

import pandas
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np


def decimal_degrees(g, m, s):
    return g + m/60 + s/3600


def plot_graph(x, y):
    df = pandas.DataFrame({
        'x_axis': x,
        'y_axis': y,
    })
    plt.plot('x_axis', 'y_axis', data=df, linestyle='-', marker='o')
    plt.show()
```

```python
# Вихідіні дані

P, I = 7, 7

x = 2 * P
y = 2 * P

H = I * 200 + P * 10
f = 100 + I * 5
h = I * 10 - P * 10

r = 10 * P - I
kyt_azumyt = (5 * I, 0, 0)

a = (0, I*10, P*10)
w = (0, -I*10, -P*10)
k = (1, I*2, P*10)
```

```python
# Градуси в радіани

def rad(g, m, s):
    return decimal_degrees(g, m, s) * pi/180
```

```python
# Задача 1.1.

rad_a = rad(*a)
rad_w = rad(*w)
rad_k = rad(*k)

print(f"а в град10 {decimal_degrees(*a)}, в радіанах {rad_a}")
print(f"w в град10 {decimal_degrees(*w)}, в радіанах {rad_w}")
print(f"k в град10 {decimal_degrees(*k)}, в радіанах {rad_k}")
```

```python
# Задача 1.2.

rad_kyt_azumyt = rad(*kyt_azumyt)

zmichenya_tochku_nahul = -(r**2/f) * sin(rad_a) * sin(rad_kyt_azumyt)
zmichenya_tochku_reilef = r * h / H

print(zmichenya_tochku_nahul)
print(zmichenya_tochku_reilef)
```

```python
# Задача 1.3.

X = H / f * x
Y = H / f * y

print(X)
print(Y)
```

```python
# Задача 1.4.

a1 = cos(rad_a)*cos(rad_k) - sin(rad_a)*sin(rad_w)*sin(rad_k)
a2 = -cos(rad_a)*sin(rad_k) - sin(rad_a)*sin(rad_w)*cos(rad_k)
a3 = -sin(rad_a) * cos(rad_w)
b1 = cos(rad_w) * sin(rad_k)
b2 = cos(rad_w) * cos(rad_k)
b3 = -sin(rad_w)
c1 = sin(rad_a)*cos(rad_k) + cos(rad_a)*sin(rad_w)*sin(rad_k)
c2 = -sin(rad_a)*sin(rad_k) + cos(rad_a)*sin(rad_w)*cos(rad_k)
c3 = cos(rad_a) * cos(rad_w)

xt = -f * (a1*x + a2*y - a3*f) / (c1*x + c2*y - c3*f)
yt = -f * (b1*x + b2*y - b3*f) / (c1*x + c2*y - c3*f)

print(xt)
print(yt)
```

```python
# Задача 1.5.

A = np.matrix(
    [
        [a1, a2, a3],
        [b1, b2, b3],
        [c1, c2, c3],
    ]
)
At = A.transpose()
l = np.matmul(A, At)

print(A)
print("\n")
print(At)
print("\n")
print(l)
```

```python
# Вихідне зображення

x_plot = [2, 1, 0, 1, 1, 2, 3, 3, 4, 3, 2]
y_plot = [0, 1, 2, 3, 2, 2, 2, 3, 2, 1, 0]

plot_graph(x_plot, y_plot)
```

```python
# Задача 2.1. Поворот площини зображення.

xy_zipped = list(zip(x_plot, y_plot))
new_x_plot = list(map(
    lambda coords: coords[0]*cos(rad_kyt_azumyt) - coords[1]*sin(rad_kyt_azumyt), 
    xy_zipped
))
new_y_plot = list(map(
    lambda coords: coords[0]*sin(rad_kyt_azumyt) + coords[1]*cos(rad_kyt_azumyt),
    xy_zipped
))


plot_graph(new_x_plot, new_y_plot)
```

```python
# Задача 2.2. Зсув зображення.

tx = P
ty = I

new_x_plot = list(map(lambda coords: coords[0] + tx, xy_zipped))
new_y_plot = list(map(lambda coords: coords[1] + ty, xy_zipped))

plot_graph(new_x_plot, new_y_plot)
```

```python
# Задача 2.3. Поворот простору.

x0 = 0.01 * P
y0 = 0.01 * I

povorot_prostor = np.matmul(A, np.array([
    [x-x0],
    [y-y0],
    [-f]
]))

print(povorot_prostor)
```

```python
# Задача 2.4. Стиснення і розтягнення зображення.

for k in (0.5, 1.5, 2):
    new_x_plot = list(map(lambda coords: k * coords[0], xy_zipped))
    new_y_plot = list(map(lambda coords: k * coords[1], xy_zipped))

    plot_graph(new_x_plot, new_y_plot)
```

```python
# Задача 2.5. Віддзеркалення відносно осей x, y.

new_y_plot = list(map(lambda coords: -coords[1], xy_zipped))

plot_graph(x_plot, new_y_plot)
```

```python
# Задача 2.6. Евклідові перетворення.

R = np.array(
    [
        [cos(rad_kyt_azumyt), -sin(rad_kyt_azumyt)],
        [sin(rad_kyt_azumyt), cos(rad_kyt_azumyt)],
    ]
)

t = np.array(
    [
        [tx],
        [ty],
    ]
)

new_coords_x = []
new_coords_y = []

for coords in xy_zipped:
    xy_matrix = np.array(
        [
            [coords[0]],
            [coords[1]],
        ]
    )
    new_coords = R * xy_matrix + t
    new_coords_x.append(new_coords[0][0])
    new_coords_y.append(new_coords[1][0])
    

plot_graph(new_coords_x, new_coords_y)

inversed_R = np.linalg.inv(R)

new_coords_zipped = list(zip(new_coords_x, new_coords_y))

old_coords_x = []
old_coords_y = []

for coords in new_coords_zipped:
    xy_matrix = np.array(
        [
            [coords[0]],
            [coords[1]],
        ]
    )
    old_coords = inversed_R*xy_matrix - inversed_R*t
    old_coords_x.append(old_coords[0][0])
    old_coords_y.append(old_coords[1][0])
    
    
plot_graph(old_coords_x, old_coords_y)
```

```python
# Задача 2.7. Проективні перетворення.

t = np.array(
    [
        [P, I, I],
        [I, P, I],
        [P, I, P],
    ]
)

new_xs = []
new_ys = []

for coords in xy_zipped:
    x = coords[0]
    y = coords[1]
    new_x = (t[0][0]*x + t[0][1]*y + t[0][2]) / (t[2][0]*x + t[2][1]*y + t[2][2])
    new_y = (t[1][0]*x + t[1][1]*y + t[1][2]) / (t[2][0]*x + t[2][1]*y + t[2][2])
    new_xs.append(new_x)
    new_ys.append(new_y)
    
plot_graph(new_xs, new_ys)
```
